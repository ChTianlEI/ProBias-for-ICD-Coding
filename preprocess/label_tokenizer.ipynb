```python
import json
import pickle
from transformers import AutoTokenizer
```

```python
tokenizer = AutoTokenizer.from_pretrained("UFNLP/gatortron-base")
desc_code = []
def_code = []
data_type = 'mimic3' # mimic4_icd9, mimic4_icd10
with open('enhanced_icd_{}.json'.format(data_type), 'rb') as f:
    labels = json.load(f)
for l in labels:
    desc_code.append(l.get('COMPREHENSIVE_DESC')+l.get('COMORBIDITY_INFO'))
    def_code.append(l.get('LONG_TITLE'))
desc_token =  tokenizer(desc_code ,return_tensors='pt', padding='max_length',truncation = True, max_length=512)
def_token =  tokenizer(def_code ,return_tensors='pt', padding='max_length',truncation = True, max_length=512)
with open("../model_data/mimic3/icd_{}_desc.pkl".format(data_type),'wb') as f:
    pickle.dump(desc_token,f)
with open("../model_data/mimic3/icd_{}_def.pkl".format(data_type),'wb') as f:
    pickle.dump(def_token,f)
```
