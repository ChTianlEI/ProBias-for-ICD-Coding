```python
import json
import pandas as pd
import numpy as np
import pickle
```

```python
with open('./icd_mimic3.json', 'r') as f:
    icd_mimic3_rank = json.load(f)
listicd = list(icd_mimic3_rank.keys())
```

```python
num_all = len(listicd)
for name in ['train', 'dev', 'test']:
    data = pd.read_csv(f'../model_data/mimic3/{name}_full.csv')
    temp_onehot = np.zeros((len(data), num_all))
    for i, row in data.iterrows():
        if isinstance(row['LABELS'], str):
            for code in row['LABELS'].split(';'):
                if len(code) == 0:
                    continue
                temp_onehot[i, listicd.index(code)] = 1
    if name == 'dev':
        name = 'val'
    np.savez(f'../model_data/mimic3/mimic3_{name}_1hot.npz', temp_onehot)
    with open(f'../model_data/mimic3/mimic3_{name}.pkl', 'wb') as file:
        pickle.dump(data, file)
```
